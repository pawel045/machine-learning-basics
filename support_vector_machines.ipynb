```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/LearnPythonWithRune/MachineLearningWithPython/main/jupyter/final/files/data_02.csv')
class_ids = {'Dobermann': 0, 'German Shepherd': 1, 'Rottweiler': 2}
colors = ['b', 'r', 'g']

data['Class_id'] = data['Class'].apply(lambda x: class_ids[x])
data['Color_class'] = data['Class_id'].apply(lambda x: colors[x])
```

```python
fig, ax = plt.subplots()
ax.scatter(x=data['Weight'], y=data['Height'], c=data['Class_id'])
```

```python
clt = SVC(kernel='linear')
clt.fit(data[['Weight', 'Height']], data['Class_id'])
```

```python
x_p = np.random.rand(10000, 2)
x_p = x_p*(40,20)+(25,55)
y_p = clt.predict(x_p)
y_p
```

```python
fig, ax = plt.subplots()
ax.scatter(x=x_p[:,0], y=x_p[:,1], c=y_p, alpha=0.25)
ax.scatter(x=data['Weight'], y=data['Height'], c=data['Color_class'])
```
